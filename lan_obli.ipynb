# Lançamento Oblíquo

### Eixo x:
  - Força viscosa $\vec{F}_{bx} = -b v_x \hat i$
$$
\sum F_x = ma_x
$$

$$
-b v_x = ma_x
$$



$$
v_x = v_{0x} e^{-bt/m}
$$

$$
v_x = v_{0}cos\theta_x e^{-bt/m}
$$


$$
x = \frac{m}{b}(v_{0}cos\theta_x)(1- e^{-kt})
$$


### Eixo y:
   - Força viscosa $\vec{F}_{by} = -b v_y \hat j$
   - Força externa: $\vec{F}_e = - \gamma z \hat j$

$$
\sum F_y = ma_y
$$

$$
-bv_y - \gamma z = ma_y
$$


$$
v_y = -\frac{\gamma z}{b} +(v_{0} cos \theta_y + \frac{\gamma}{b}) e^{-bt/m}
$$

$$
y = -\frac{\gamma z}{b}t +\frac{m}{b}(v_{0} cos \theta_y + \frac{\gamma}{b}) (1-e^{-bt/m})
$$

### Eixo z:
  - Força viscosa $\vec{F}_{bz} = -b v_z \hat k$
  - Aceleração da gravidade: $\vec{g} = -g \hat k$

$$
\sum F_z = ma_z
$$

$$
-bv_z - mg = ma_z
$$


$$
v_z = -\frac{mg}{b} +(v_{0} cos \theta_z + \frac{mg}{b}) e^{-bt/m}
$$

$$
z =  -\frac{mg}{b}t + \frac{m}{b}(v_0 cos\theta_z + \frac{mg}{b})(1-e^{-bt/m})
$$
### Velocidade lançamento:

 $\vec v = v_0 cos \theta_x \hat i + v_0 cos \theta_y \hat j + v_0 cos \theta_z \hat k$ 





```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 8, 100)
v0 = 100.0
theta_x = np.pi/3
theta_y = np.pi/3
theta_z = np.pi/4

b = 1.0
gamma = 1.0
g = 9.81
m = 1.0

k = b/m

vx = (v0*np.cos(theta_x))*np.exp(-k*t)
vy = -(gamma*z/b) + (v0*np.cos(theta_y)+(gamma/k))*np.exp(-k*t)
vz = -(g/k) + (v0*np.cos(theta_z)+(g/k))*np.exp(-k*t)


z = -(g/k)*t + (1/k)*(v0*np.cos(theta_z)+(g/k))*(1-np.exp(-k*t))
x = (v0*np.cos(theta_x))/k*(1-np.exp(-k*t)) 
y = -(gamma*z/b)*t + (1/k)*(v0*np.cos(theta_y)+(gamma/k))*(1-np.exp(-k*t))

```

```python
#vx versus t
#Trajetória no plano xz
plt.plot(t,vx)
plt.grid(True)
```

```python
#vx, vt e vz
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(vx, vy, vz)
plt.grid(True)

plt.show()
```
